# Analysing sound wave using fast Fourier transform



```python
import matplotlib.pyplot as plt
import numpy as np

Np = 1000 # number of particles
Lx = 10
Ly = 1
Dx = 0.25 # maximum displacement
lamb = 1.0 # wavelength
phi = 0.0 # phase difference

x = np.random.uniform(0, Lx, Np)
y = np.random.uniform(0, Ly, Np)
x = x + Dx*np.sin(np.pi*x/lamb)

fig, ax = plt.subplots(3, figsize=(10,7))

for n in range(0, 3, 1):
    ax[n].set_title('$t = 0$', fontsize=14)
    ax[n].set_xlabel('x', fontsize=12)
    ax[n].set_ylabel('y', fontsize=12)
    ax[n].set_xlim(0, Lx)
    ax[n].set_ylim(0, Ly)
    ax[n].set_aspect(1)
    ax[n].tick_params(axis='both', which='major', labelsize=12)
    ax[n].set_yticks(np.arange(0, 2, 1))
    ax[n].set_xticks(np.arange(0, 11, 1))
    ax[n].scatter(x, y, s=2)

plt.show()
```
